```python
from __future__ import print_function
import time
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
init=pd.read_csv('pythonDS.csv',sep=';')
```

```python
init
```

```python
second = init[['Intelligence','Strength','Speed','Durability','Power','Combat']]
```

```python
second
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(second)


tsne_results[:,0]
```

```python
third = init[['Alignment','Intelligence','Strength','Speed','Durability','Power','Combat']]

third['tsne-one'] = tsne_results[:,0]
third['tsne-two'] = tsne_results[:,1]

g =sns.scatterplot(x="tsne-one", y="tsne-two",            hue="Alignment",          data=third)
g

plt.savefig('Python_TSNE.png') 
```
